import numpy as np
import matplotlib.pyplot as plt

# Problem Statement 
I will use a motivating example of housing price prediction with simple data set with only two data points - a house with 1000 square feet(sqft) sold for \\$300,000 and a house with 2000 square feet sold for \\$500,000.. These two points will constitute our data or training set.Tthe units of size are 1000 sqft and the units of price are 1000s of dollars

| Size (1000 sqft) | Price (1000s of dollars) |
|------------------|--------------------------|
| 1.0              | 300                      |
| 2.0              | 500                      |.

```python
# X_train is the input variable 
# y_train is the output variable 
X_train = np.array([1.0, 2.0])
y_train = np.array([300, 500])
print(f'X_train = {X_train}')
print(f'y_train = {y_train}')
```

# Number of Training Examples `m`
`m` to denotes the number of training examples. Numpy arrays have a `.shape` parameter. `x_train.shape` returns a python tuple with an entry for each dimension. `x_train.shape[0]` is the length of the array.

```python
print(f'X_train.shape = {X_train.shape}')
m = X_train.shape[0]
print(f'The Number of Training Examples = {m}')
```

# Training Example 
(x$^{(i)}$, y$^{(i)}$) denotes the ${i ^{th}}$ training example, (x$^{(0)}$, y$^{(0)}$) is the first training example.

```python
i = 0

x_i = X_train[i]
y_i = y_train[i]
print(f'(x^({i}), y^({i})) = {x_i, y_i}')
```

# Plotting The Data 
We can use `scatter()` function in `matplotlib` to plot the two points.

```python
plt.scatter(X_train, y_train, marker = 'X', c= 'r')
plt.xlabel('Size of The House')
plt.ylabel('Price of The House')
plt.title('Housing Prices')
```

# Model Function
The model function for linear regression (which is a function that maps from x to y) is represented as $f_{w, b} = wx + b$, which is a straight line equation and different values of $w$ and $b$ gives different straight lines.

let's try $w = 100$ and $b = 100$

```python
def compute_model_output(x, w, b):
    m = x.shape[0]
    f_wb = []
    for i in range(m):
        f_wb.append(w * x[i] + b)
    return f_wb
```

```python
w = 200
b = 100
tmp_f_wb = compute_model_output(X_train, w, b)
```

```python
# plotting the actual values
plt.scatter(X_train, y_train, marker = 'X', c = 'r', label = 'Actual Data')

# plotting the model output
plt.plot(X_train, tmp_f_wb, marker = 'X', c = 'b', label = 'Predicted Data')

# set the title
plt.title('Housing Prices')

# set the x-axis label
plt.xlabel('Size of The House')

# set the y-axis label
plt.ylabel('Price of The House')

plt.legend()
plt.show()
```

We can see that $w = 100$ $b = 100$ does not fit our data, so we can try some values and the optimum is $w = 200$ and $b = 100$ 


# Prediction 
Let's predict the price of a house with 1200 sqft. Since the units of $x$ are in 1000's of sqft, $x$ is 1.2.

```python
w = 200
b = 100
x_i = 1.2
# f_wb = w*x + b
price = 200 * 1.2 + 100
print(f'The Price of 1200 sqft = {price}')
```

```python
w = 200
b = 100
tmp_f_wb = compute_model_output(X_train, w, b)

# plotting the actual values
plt.scatter(X_train, y_train, marker = 'X', c = 'r', label = 'Actual Data')

# plotting the model output
plt.plot(X_train, tmp_f_wb, marker = 'X', c = 'b', label = 'Predicted Data')

# set the title
plt.title('Housing Prices')

# set the x-axis label
plt.xlabel('Size of The House')

# set the y-axis label
plt.ylabel('Price of The House')

plt.scatter(x_i, price, marker = 'X', c = 'g')

plt.legend()
plt.show()
```
